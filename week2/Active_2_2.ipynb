ผู้เล่นกำลังเดินทางผ่านป่าลึกลับและพบทางแยก 2 ทาง ผู้เล่นต้องเลือกเส้นทางเพื่อเดินหน้าต่อ แต่ละเส้นทางจะนำไปสู่เหตุการณ์และผลลัพธ์ที่แตกต่างกัน

```python
print("=== เกมผจญภัยในป่าลึกลับ ===")
print("คุณพบทางแยก 2 ทาง")
print("1. ทางซ้าย")
print("2. ทางขวา")

choice = input("เลือกเส้นทาง (1 หรือ 2): ")

if choice == "1":
    print("\nคุณพบหีบสมบัติ")
    print("1. เปิดหีบสมบัติ")
    print("2. เดินผ่านไป")

    action = input("เลือก (1 หรือ 2): ")

    if action == "1":
        print("\nคุณเปิดหีบสมบัติ")
        print("คุณได้เจอผีเฝ้าสมบัติ 67 ตามหลอก")
        print("คุณตายเพราะผี 67")
    elif action == "2":
        print("\nคุณเดินผ่านหีบสมบัติ")
        print("คุณออกเดินมาเจอผีทุงทุงทุงวิ่งไล่จนออกป่าได้")
    else:
        print("\nกรุณาเลือก 1 หรือ 2")

elif choice == "2":
    print("\nคุณพบมังกรเฝ้าถ้ำ")
    print("1. ต่อสู้กับมังกรด้วยท่า 67 ")
    print("2. วิ่งหนี")

    action = input("เลือก (1 หรือ 2): ")

    if action == "1":
        print("\nคุณต่อสู้กับมังกร")
        print("คุณถูกมังกรโจมตีด้วยท่า 123")
        print("เกมจบ")
    elif action == "2":
        print("\nคุณวิ่งหนีจากมังกร")
        print("คุณหนีไม่สำเร็จเพราะผีทุงทุงดักหน้า")
    else:
        print("\nกรุณาเลือก 1 หรือ 2")

else:
    print("\nกรุณาเลือกเฉพาะ 1 หรือ 2")
```
