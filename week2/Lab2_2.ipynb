```python
# รับอายุจากผู้ใช้
age = int(input("กรุณากรอกอายุ: "))

# แนะนำเรตภาพยนตร์
if age < 13:
    print("แนะนำเรต G")
elif age < 15:
    print("แนะนำเรต PG")
elif age < 18:
    print("แนะนำเรต 15+")
else:
    print("แนะนำเรต 18+")

# Challenge
guardian = input("มีผู้ปกครองมาด้วยหรือไม่ (yes/no): ")

if age < 18 and guardian == "yes" or guardian == "y" :
    print("สามารถรับชมได้เมื่อมีผู้ปกครอง")
elif age >= 18:
    print("สามารถรับชมได้")
else:
    print("ไม่สามารถรับชมได้")
```
