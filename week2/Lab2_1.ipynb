```python
num = int(input("input :"))

#ตรวจสอบว่าเป็นจำนวนอะไรก่อน
if num > 0:
    x = "จำนวนบวก"
elif num < 0:
    x = "จำนวนลบ"
else :
    x = "จำนวนศูนย์"

#ตรวจสอบว่าเป็นเลขคู่หรือคี่
if num % 2 == 0:
    y = "คู่"
else :
    y = "คี่"

print("แสดงผล")
print(x)
print(y)


```
